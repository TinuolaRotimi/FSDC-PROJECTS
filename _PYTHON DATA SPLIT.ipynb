```python
new_add=data['Address'].str.split(r'[,\s-]+',expand=True)
new_add
```

```python
new_add.columns=[['House_No','BlocKNo','Area','City','Zipcode']]
new_add
```

```python
data.head()
```

```python
new_date=data['Date'].str.split(r'[-]',expand=True)
```

```python
new_date
```

```python
new_time=data['Time'].str.split(r'[\s]',expand=True)
new_time
```

```python
new_time.columns=[['Hour','Minutes','Seconds']]
new_time
```

```python
data.head()
```

```python
new_data=pd.concat([new_add,new_date,new_time], axis=1)
new_data
```

```python
new_data.to_excel('output.xlsx', index=False)
```
